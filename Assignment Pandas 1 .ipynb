4.Create Function that take dataset columns and replace whitespace with Underscore and update dataset automatically and the pass the dataset.

```python
def replace_whitespace_with_underscore(data):
    data.columns = data.columns.str.replace(' ', '_')
    return data

data = replace_whitespace_with_underscore(data)

print(data.head())
```

10.Find out the string in Deaths Column and replace it with the mean of Deaths column

```python
data['Deaths'] = pd.to_numeric(data['Deaths'], errors='coerce')
mean_deaths = data['Deaths'].mean()
data['Deaths'].fillna(mean_deaths, inplace=True)
data['Deaths']
```

18. Make new column name Total_cases that have a sum of Deaths, Recovered , Active

```python
data['Deaths'].fillna(0, inplace=True)
data['Deaths']
data['Recovered'].fillna(0, inplace=True)
data['Recovered']
data['Active'].fillna(0, inplace=True)
data['Active']
```

```python
data['Total'] = data['Deaths'] + data['Recovered'] + data['Active']
data['Total']
```

19.
Calculate Death_Rate for Each Country (Deaths per Total Cases) by suing formula (Desths/Total_cases)*100 and save it in Dath_Rate column

```python
data['Death_Rate'] = (data['Deaths'] / data['Total']) * 100

print(data[['Country/Region', 'Deaths', 'Total', 'Death_Rate']])
```

20.
Identify countries where total cases are increasing but death rates remain low.

```python
death_rate_threshold = 1.0  
week_change_threshold = 0  

filtered_data = data[(data['1_week_change'] > week_change_threshold) & (data['Deaths_/_100_Cases'] < death_rate_threshold)]

print(filtered_data[['Country/Region', '1_week_change', 'Deaths_/_100_Cases', 'Death_Rate']])
```
