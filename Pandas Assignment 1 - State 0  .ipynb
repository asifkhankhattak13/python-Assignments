```python
import pandas as pd
```

```python
state0 = pd.read_csv('states0.csv').drop("Unnamed: 0", axis = 1) 
```

```python
state0
```

```python
state0.Hispanic = state0.Hispanic.str.replace("%"," "). astype(float)*100
```

```python
state0.dtypes
```

```python
state0.Hispanic
```

```python
state0.White = state0.White.str.replace("%"," ").astype(float)*100
```

```python
state0.Black = state0.Black.str.replace("%"," "). astype(float)*100
state0.Native = state0.Native.str.replace("%"," "). astype(float)*100
state0.Asian = state0.Asian.str.replace("%"," "). astype(float)*100
state0.Pacific = state0.Pacific.str.replace("%"," "). astype(float)*100
```

```python
state0.Income = state0.Income.str.replace("$"," "). astype(float)
```

```python
state0
```

```python
state0["males"], state0["female"] = state0.GenderPop.str.split("_").str[0],state0.GenderPop.str.split("_").str[1]
```

```python
state0["males"] = state0["males"].str.replace("M", "").astype(int) 
```

```python
state0["females"] = state0["female"].str.replace("F", "").astype(int) 
```

```python
state0
```

```python
state0.drop(columns=["GenderPop"], inplace=True)
```

```python
state0
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(state0)

# Create scatter plot
plt.scatter(df["females"], df["Income"])

# Add labels and title
plt.xlabel("Females")
plt.ylabel("Income")
plt.title("Income/ Females")

# Show the plot
plt.show()
```
