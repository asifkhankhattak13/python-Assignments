```python
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Download the vader lexicon for sentiment analysis if not already present 
nltk.download('vader_lexicon')
# the vader lexicon is a list of words and present 
# rating that is used in , polarity scores are used in sentiment athe vader ( Valence Aware Dictionary and sentiment )
```

```python
#  A polarity score is a numerical rating that measures the overall tone of text , phrase or word and indicates 
#whether it is positive negative or nueutral 
# polarity scores are used for in sentiments analysis which is 
```

```python
sia = SentimentIntensityAnalyzer()

def calculate_sentiment(review):
    sentiment = sia.polarity_scores(review)
    print(sentiment)
    return sentiment['compound']
    
```

```python
## test code

review = 'this is a great product '
print(calculate_sentiment(review))
```

```python
reviews = pd.DataFrame(reviews)
reviews
```

```python
reviews['SentimentScore'] = reviews['ReviewText'].apply(calculate_sentiment)
```

```python
#print(reviews)
```

```python
# define a function to categories sentiments using both sentiment score and rating 
def categorize_sentiment(score, rating):
    # use both score and rating to determine 
    if score > 0.05 : #positve
        if rating >= 4:
            return 'Positive'
        elif rating == 3:
            return 'Mixed Positive'
        else:
            return 'Mixed Negative'
    elif score < -0.05:
        if rating <= 2:
            return 'Negative'
        elif rating == 3:
            return 'Mixed Negative'
        else:
            return 'Mixed Positive'
    else:
        if rating >= 4:
            return 'Positive'
        elif rating <= 2:
            return 'Negative'
        else:
            return 'Neutral'
```

```python
#apply sentiment 
reviews['SentimentCategory'] = reviews.apply(lambda row:categorize_sentiment(row['SentimentScore'], row['Rating']), axis=1)
reviews
```

```python
# creating a category so that continous value to discrete so that we can plot a grapgh
```

```python
def sentiment_bucket(score):
    if score >= 0.5:
        return '0.5 to 1'
    elif 0.0 <= score < 0.5 :
        return '0.0 to 0.49'
    elif -0.5 <= score <0.0:
        return '-0.49 to 0.0'
    else:
        return '-1.0 to -0.5'
```

```python
reviews['SentimentBucket'] = reviews['SentimentScore'].apply(sentiment_bucket)
reviews
```
